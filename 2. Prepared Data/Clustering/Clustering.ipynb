```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
#mouth
open_mouth_ds = pd.read_csv('../../1. Original Data/Person1/open_mouth.csv', delimiter = ';', header = None)
smile_ds = pd.read_csv('../../1. Original Data/Person1/smile.csv', delimiter = ';', header = None)
kiss_ds = pd.read_csv('../../1. Original Data/Person1/kiss.csv', delimiter = ';', header = None)
pl_ds = pd.read_csv('../../1. Original Data/Person1/purse_lips.csv', delimiter = ';', header = None)


```

```python
#each row represents one frame
#the columns have the position information in x,y pairs e.g col 0 is the x coordinate for the first landmark, col1 is the y coordinate
def pos_time (dataset, index):
    index = index-1
    x = dataset.iloc[:, index*2:index*2+2].values
    return x

#all measurements are normalised against the distance between the eyes
#vertical length of the mouth

def distance(mouth_ds, p1, p2):
    v1 = pos_time(mouth_ds, p1)
    v2 = pos_time(mouth_ds, p2)
    n1 = pos_time(mouth_ds, 40)
    n2 = pos_time(mouth_ds, 43)
    ret = []
    for i in range(len(v1)):
        v = (abs(math.dist(v1[i], v2[i])))
        n = (abs(math.dist(n1[i], n2[i])))
        ret.append(v/n)
    return ret

def mouth_vlen(mouth_ds):
    return distance(mouth_ds, 52,58)

#horizontal length of the mouth
def mouth_hlen(mouth_ds):
    return distance(mouth_ds, 55, 49)

allX = []
allY = []
datasets = [open_mouth_ds, smile_ds, kiss_ds, pl_ds]
for ds in datasets:
    x = mouth_hlen(ds)
    y = mouth_vlen(ds)
    allX.extend(x)
    allY.extend(y)
        
X = np.column_stack((allX, allY))
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
all_datasets = pd.concat([open_mouth_ds, smile_ds, kiss_ds, pl_ds], ignore_index=True)
Xall = all_datasets.iloc[:,:].values
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(Xall)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
